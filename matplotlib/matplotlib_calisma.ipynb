```python
data2 = pd.read_csv("usd_jpy.csv",parse_dates=["date"])
```

```python
data2.info()
```

```python
fig,ax = plt.subplots()
ax.plot(data2["date"],data2["open"])
ax.set_xlabel("Time")
ax.set_ylabel("Open")
plt.show()
```

```python
data2.set_index("date",inplace=True)
```

```python
data2.index
```

```python
millenium = data2["2000-07-13":"2000-12-31"]
millenium
```

```python
fig,ax = plt.subplots()
ax.plot(millenium.index,millenium["open"])
ax.set_xlabel("Time")
ax.set_ylabel("Open")
ax.set_title("USD/JPY in year 2000")
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(millenium.index,millenium["open"])
ax.set_xlabel("Time")
ax.set_ylabel("Open")
ax.set_title("USD/JPY in year 2000")

ax2 = ax.twinx()
ax2.plot(millenium.index,millenium["close"],
         color="yellow",linestyle="--")
ax2.set_ylabel("Close")
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(millenium.index,millenium["open"])
ax.set_xlabel("Time")
ax.set_ylabel("Open", color="blue")
ax.set_title("USD/JPY in year 2000")
ax.tick_params("y",colors="blue") #changes the ticks color
ax2 = ax.twinx()
ax2.plot(millenium.index,millenium["close"],
         color="yellow",linestyle="--")
ax2.set_ylabel("Close",color="red")
ax2.tick_params("y",color="red") 
#color vs colors.
# color changes only ticks and label name
# colors changes ticks and numbers
plt.show()
```

```python
medals = pd.read_csv("data/medals_by_country_2016.csv",index_col=0)
```

```python
medals.head()
```

```python
fig, ax = plt.subplots()
ax.bar(medals.index,medals["Gold"])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medals.index,medals["Gold"])
ax.set_xticklabels(medals.index,rotation=90)
ax.set_ylabel("Number of Medals")
plt.show()
```

```python
fig,ax = plt.subplots()
ax.bar(medals.index,medals["Gold"])
ax.bar(medals.index,medals["Silver"],bottom=medals["Gold"])
ax.set_xticklabels(medals.index,rotation = 90)
ax.set_ylabel("Number of Medals")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medals.index,medals["Gold"])
ax.bar(medals.index,medals["Silver"],bottom=medals["Gold"])
ax.bar(medals.index,medals["Bronze"],bottom=medals["Gold"]+medals["Silver"])
ax.set_xticklabels(medals.index,rotation=90)
ax.set_ylabel("Number of Medals")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medals.index,medals["Gold"],label="Gold")
ax.bar(medals.index,medals["Silver"],
       bottom=medals["Gold"],label="Silver")
ax.bar(medals.index,medals["Bronze"],
       bottom=medals["Gold"]+medals["Silver"],label="Bronze")
ax.set_ylabel("Number of Medals")
ax.set_xticklabels(medals.index,rotation=90)
ax.legend()
plt.show()
```
